```python
import pandas as pd

medals = pd.read_csv('/content/medals.csv')
medals.head()
```

```python
cntry = medals['Unnamed: 0'][:5]
bronze = medals['Bronze'][:5]
gold = medals['Gold'][:5]
silver = medals['Silver'][:5]
```

```python
import matplotlib.pyplot as plt

fig,ax = plt.subplots()
ax.bar(cntry,bronze)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.barh(cntry,bronze)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.barh(cntry,bronze,color='red')
plt.show()
```

```python
fig,ax = plt.subplots()
ax.barh(cntry,bronze,color='red',edgecolor='blue')
plt.show()
```

```python
fig,ax = plt.subplots()
ax.bar(cntry,bronze)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.bar(cntry,bronze)
ax.bar(cntry,gold,bottom=bronze)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.bar(cntry,bronze)
ax.bar(cntry,gold,bottom=bronze)
ax.bar(cntry,silver,bottom=bronze+gold)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(cntry,bronze,width=0.5)
ax.bar(cntry,gold,width=0.5)
plt.show()
```

```python
import numpy as np
x = np.arange(len(cntry))

width = 0.3

fig, ax = plt.subplots()
ax.bar(x-0.3,bronze,width=width)
ax.bar(x,gold,width=width)
ax.bar(x+0.3,silver,width=width)
plt.show()
```

# Pie Chart

```python
medals['totals'] = medals['Bronze'] + medals['Silver'] + medals['Gold']
totals = medals['totals'][:5]

fig,ax = plt.subplots()
ax.pie(bronze)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.pie(bronze,labels=cntry)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.pie(bronze,labels=cntry,autopct='%1.1d%%')
plt.show()
```

```python
fig,ax = plt.subplots()
ax.pie(bronze,labels=cntry,autopct='%1.1f%%')
plt.show()
```

```python
fig,ax = plt.subplots()
ax.pie(bronze,labels=cntry,autopct='%1.2f%%')
plt.show()
```

```python
plt.subplots(figsize=(8,8))
my_circle = plt.Circle((0, 0), 0.7, color='white')
plt.pie(totals, labels=cntry, autopct='%1.1f%%',
            startangle=90, labeldistance=1.05)
plt.gca().add_artist(my_circle)
            
plt.show()
```

```python
plt.subplots(figsize=(8,8))
my_circle = plt.Circle((0, 0), 0.7, color='red')
plt.pie(bronze, labels=cntry, autopct='%1.1f%%',
            startangle=90, labeldistance=1.05)
plt.gca().add_artist(my_circle)
            
plt.show()
```

```python
plt.subplots(figsize=(8,8))
my_circle = plt.Circle((-1, 0), 0.7, color='red')
plt.pie(bronze, labels=cntry, autopct='%1.1f%%',
            startangle=90, labeldistance=1.05)
plt.gca().add_artist(my_circle)
            
plt.show()
```

```python
plt.subplots(figsize=(8,8))
my_circle = plt.Circle((-1, 0), 0.1, color='red')
plt.pie(bronze, labels=cntry, autopct='%1.1f%%',
            startangle=90, labeldistance=1.05)
plt.gca().add_artist(my_circle)
            
plt.show()
```

# Figure Set

```python
fig,ax = plt.subplots()
ax.bar(cntry,bronze,label='Bronze')
ax.bar(cntry,gold,bottom=bronze,label='Gold')
ax.bar(cntry,silver,bottom=bronze+gold,label='Silver')
ax.legend()
plt.show()
```

```python
fig,ax = plt.subplots()
ax.bar(cntry,bronze,label='Bronze')
ax.bar(cntry,gold,bottom=bronze,label='Gold')
ax.bar(cntry,silver,bottom=bronze+gold,label='Silver')
ax.legend()
ax.set_title('Medal')
plt.show()
```

```python
fig,ax = plt.subplots()
ax.bar(cntry,bronze,label='Bronze')
ax.bar(cntry,gold,bottom=bronze,label='Gold')
ax.bar(cntry,silver,bottom=bronze+gold,label='Silver')
ax.legend()
ax.set_title('Medal')
ax.set_xlabel('Country')
ax.set_ylabel('Total Medals')
plt.show()
```
